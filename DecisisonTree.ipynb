```python
import numpy as np
import pandas as pd
import tensorflow_decision_forests as tfdf
```

```python
path = "https://storage.googleapis.com/download.tensorflow.org/data/palmer_penguins/penguins.csv"
df = pd.read_csv(path)
```

```python
df
```

```python
labels = "species"
print(df["species"].unique())
```

```python
percentage = np.random.rand(len(df)) <0.11
```

```python
true_list = []
for i in range(len(percentage)):
  if percentage[i] == True:
    true_list.append(i)
  else:
    continue
```

```python
true_list
```

```python
len(true_list)
```

```python
train_data_set = df[~percentage]
test_data_set = df[percentage]
```

```python
test_data_set
```

```python
df_train = tfdf.keras.pd_dataframe_to_tf_dataset(train_data_set, label = labels)
model = tfdf.keras.CartModel(max_depth = 10)
model.fit(df_train)
```

```python
tfdf.model_plotter.plot_model_in_colab(model)
```

```python
model.compile("accuracy")
```

```python
df_test = tfdf.keras.pd_dataframe_to_tf_dataset(test_data_set, label = labels)
model.evaluate(df_test)
```
