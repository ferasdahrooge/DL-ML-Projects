```python
import tensorflow as tf
import numpy as np
```

```python
celsius_degrees = np.array([10,23,-1,100,-10,60,-4,0,-40,90,-11.1111], dtype= float)
fehrenhiet_degree = np.array([50,73.4,30.2,212,14,140,24.8,32,-40,194,12], dtype = float)

for i, c in enumerate(celsius_degrees):
  print(i, " this is I")
  print(c, " this is C")
```

```python
layer_0 = tf.keras.layers.Dense(units = 1, input_shape = [1])
```

```python
model = tf.keras.Sequential([layer_0])
model1 = tf.keras.Sequential([layer_0])
model2 = tf.keras.Sequential([layer_0])
model3 = tf.keras.Sequential([layer_0])
model4 = tf.keras.Sequential([layer_0])
model5 = tf.keras.Sequential([layer_0])
print("models are built")
```

```python
model.compile(loss = "mean_squared_error",
                    optimizer = tf.keras.optimizers.Adagrad(0.1))
model1.compile(loss = "mean_squared_error",
                    optimizer = tf.keras.optimizers.Ftrl(0.1))
model2.compile(loss = "mean_squared_error",
                    optimizer = tf.keras.optimizers.Nadam(0.1))
model3.compile(loss = "mean_squared_error",
                    optimizer = tf.keras.optimizers.Adamax(0.1))
model4.compile(loss = "mean_squared_error",
                    optimizer = tf.keras.optimizers.Adam(0.1))
model5.compile(loss = "mean_squared_error",
                    optimizer = tf.keras.optimizers.Adadelta(0.1))
```

```python
train1 = model.fit(celsius_degrees, fehrenhiet_degree, epochs =500, verbose = False)
train2 = model1.fit(celsius_degrees, fehrenhiet_degree, epochs =500, verbose = False)
train3 = model2.fit(celsius_degrees, fehrenhiet_degree, epochs =500, verbose = False)
train4 = model3.fit(celsius_degrees, fehrenhiet_degree, epochs =500, verbose = False)
train5 = model4.fit(celsius_degrees, fehrenhiet_degree, epochs =500, verbose = False)
train6 = model5.fit(celsius_degrees, fehrenhiet_degree, epochs =500, verbose = False)
```

```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
figure(figsize=(10, 100), dpi=80)

plt.xlabel("Epoch Number")
plt.xlabel("Loss Magnitude")
plt.plot(train1.history["loss"], label = "Adagrad")
plt.plot(train2.history["loss"], label = "Ftrl")
plt.plot(train3.history["loss"], label = "Nadam")
plt.plot(train4.history["loss"], label = "Adamax")
plt.plot(train5.history["loss"], label = "Adam")
plt.plot(train6.history["loss"], label = "Adadelta")
plt.legend(loc="lower left")
plt.show()
```

```python
print("predictions regarding Adagrad: ", model.predict([42]))
print("predictions regarding Ftrl: ", model1.predict([42]))
print("predictions regarding Nadam: ", model2.predict([42]))
print("predictions regarding Adamax: ", model3.predict([42]))
print("predictions regarding Adam: ", model4.predict([42]))
print("predictions regarding Adadelta: ", model5.predict([42]))
```
